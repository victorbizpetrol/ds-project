### Data Segmentation
Segmenting banks by asset size:  small, med, & large banks.

```python
#Small <$50M, Medium = $50M to $50B, Large = $50B to $3T
car_df['Bank Size'] = pd.cut(car_df['Total Assets'], [0,50000, 50000000, 3000000000], 
labels=['Small', 'Medium', 'Large'])
```
