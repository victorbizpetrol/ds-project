### **Importing csv files into dataframes**

```python
#Banks Assets and Liabilities csv
al_df = pd.read_csv('Data/Bank_Assets_and_Liabilities.csv')
```

```python
#Drop columns
al_df.drop(['docket', 'rssdhcr', 'rundate', 'bkclass', 'offfor', 'stmult', 'specgrp', 'subchaps', 'county', 'cbsa_metro',
            'insdate', 'effdate', 'parcert', 'trust', 'regagnt', 'insagnt1', 'fdicdbs', 'fdicsupv', 'fldoff',
            'fed', 'occdist', 'otsregnm', 'offoa', 'cb', 'webaddr'
           ], axis=1, inplace=True)
```

```python
al_df.head()
```

```python
#Banks Performance Ratios csv
pr_df = pd.read_csv('Data/Bank_Performance_Ratios.csv')
```

```python
#Drop columns
pr_df.drop(['docket', 'rssdhcr', 'name','city', 'stalp', 'zip', 'repdte', 'rundate', 'bkclass', 'address', 'namehcr', 'offdom',
            'offfor', 'stmult', 'specgrp', 'subchaps', 'county', 'cbsa_metro', 'insdate', 'effdate', 'parcert', 'trust', 'regagnt', 
            'insagnt1', 'fdicdbs', 'fdicsupv', 'fldoff', 'fed', 'occdist', 'otsregnm', 'offoa', 'cb', 'webaddr'
           ], axis=1, inplace=True)
```

```python
pr_df.head()
```

```python
#Banks Income and Expense csv
ie_df = pd.read_csv('Data/Bank_Income_ and_Expense.csv')
```

```python
#Drop columns
ie_df.drop(['docket', 'rssdhcr', 'name','city', 'stalp', 'zip', 'repdte', 'rundate', 'bkclass', 'address', 'namehcr', 'offdom',
            'offfor', 'stmult', 'specgrp', 'subchaps', 'county', 'cbsa_metro', 'insdate', 'effdate', 'parcert', 'trust', 'regagnt', 
            'insagnt1', 'fdicdbs', 'fdicsupv', 'fldoff', 'fed', 'occdist', 'otsregnm', 'offoa', 'cb', 'inst.webaddr'
           ], axis=1, inplace=True)
```

```python
ie_df.head()
```

```python
merge_alpr_df = pd.merge(al_df, pr_df, on='fed_rssd')
```

```python
car_df = pd.merge(merge_alpr_df, ie_df, on='fed_rssd')
```

```python
car_df.head()
```

```python
car_df.to_csv('Data/test.csv')
```
